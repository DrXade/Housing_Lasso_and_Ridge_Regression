```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
X= housing.data
y= housing.target
```

```python
X.shape
```

```python
housing.target_names
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=.20, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
X_Scaler = StandardScaler()
X_train = X_Scaler.fit_transform(X_train)
X_test = X_Scaler.transform(X_test)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso()
lasso_reg.fit(X_train,y_train)
```

```python
y_pred_lasso_reg = lasso_reg.predict(X_test)
```

```python
print(mean_absolute_error(y_test,y_pred_lasso_reg))
```

```python
print(mean_squared_error(y_test,y_pred_lasso_reg))
```

```python
print(r2_score(y_test,y_pred_lasso_reg))
```

```python
param_grid = { 

        'alpha' : [.00155, .0016, .00162,.00163,.00164]

}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lasso_cv=GridSearchCV(lasso_reg, param_grid, cv=3, n_jobs =-1) 
```

```python
lasso_cv.fit(X_train,y_train)
```

```python
y_pred_lasso_cv = lasso_cv.predict(X_test)
```

```python
print(mean_absolute_error(y_test,y_pred_lasso_cv))
```

```python
print(mean_absolute_error(y_test,y_pred_lasso_cv))
```

```python
print(r2_score(y_test,y_pred_lasso_cv))
```

```python
from sklearn.linear_model import Ridge
Ridge_reg = Ridge()
```

```python
Ridge_reg.fit(X_train,y_train)
```

```python
ridge_cv=GridSearchCV(Ridge_reg, param_grid, cv=3, n_jobs =-1) 
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
y_pred_ridge_cv = ridge_cv.predict(X_test)
```

```python
print(mean_absolute_error(y_test,y_pred_ridge_cv))
```

```python
print(mean_absolute_error(y_test,y_pred_ridge_cv))
```

```python
print(r2_score(y_test,y_pred_ridge_cv))
```
